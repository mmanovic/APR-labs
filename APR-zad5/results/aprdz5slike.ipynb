```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt

x=np.linspace(0,1,500)
with open("zad4trapeseStabilno.txt", 'r') as f:
    lines = []
    f1=[]
    f2=[]
    for i in f:
        lines.append(i.rstrip())
    for line in lines:
        tmp = line.split(' ')
        f1.append(float(tmp[0]))
        f2.append(float(tmp[1]))
    plt.figure(figsize=(14,7))
    plt.plot(x,f1,label="x1")
    plt.plot(x,f2,label="x2")
    plt.legend()
    plt.show()
```
